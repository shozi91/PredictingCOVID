```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 500)
```

```python
covid_df = pd.read_csv('03-16-2020.csv')
#number of countries
covid_df['Country/Region'].nunique()
#turn last update to datetime
#to_datetime method. exclude time. just keep date. use map - string slicing method
print(covid_df.dtypes)
covid_df
```

```python
world_pop_df = pd.read_csv('world_countries_gdp.csv')
#number of countries
world_pop_df['country'].nunique()
world_pop_df
```

```python
#get list of countries from datasets
mod_covid_df = covid_df.copy()
mod_covid_df.drop(mod_covid_df.columns.difference(['Country/Region']), 1, inplace=True)
mod_covid_df.replace("US","United States", inplace=True)
mod_covid_df.reset_index(drop=True,inplace=True)

mod_world_pop_df = world_pop_df.copy()
mod_world_pop_df.drop(mod_world_pop_df.columns.difference(['country']), 1, inplace=True)
mod_world_pop_df.rename(columns = {'country': "Country/Region"}, inplace=True)
mod_world_pop_df.reset_index(drop=True,inplace=True)
```

```python
countries_list = []
#get number of countries in both datasets
for covid_index,covid_row in mod_covid_df.iterrows():
    for world_pop_df_index,world_pop_df_row in mod_world_pop_df.iterrows():
        if covid_row[0] == world_pop_df_row[0]:
            countries_list.append(covid_row[0])
len(set(countries_list))
```

```python
grouped_covid_df = covid_df.groupby('Country/Region').sum()
grouped_covid_df.drop(['Latitude','Longitude'],1,inplace=True)
grouped_covid_df.rename(index = {"US":"United States","Taiwan*":"Taiwan"}, inplace=True)
grouped_covid_df.reset_index(inplace=True)
grouped_covid_df.rename(columns = {"Country/Region":"Country"}, inplace=True)
#grouped_covid_df.query('Country_Region.str.contains("Congo")',engine='python',inplace=True)
```

```python
grouped_world_pop_df = world_pop_df.groupby('country').sum()
grouped_world_pop_df.drop(['rank','unGDP'],1,inplace=True)
grouped_world_pop_df.rename(columns = {"imfGDP": "GDP","gdpPerCapita": "GDP Per Capita","pop":"Population"},inplace=True)
grouped_world_pop_df.index.name = 'Country'
grouped_world_pop_df.reset_index(inplace=True)
```

```python
merged_df = grouped_covid_df.merge(grouped_world_pop_df,how='left',on='Country')
merged_df.dropna(inplace=True)
merged_df.reset_index(drop=True,inplace=True)
merged_df['GDP Per Capita'] = merged_df['GDP Per Capita'].map('{:.0f}'.format)
merged_df['Population'] = (merged_df['Population'] * 1000).astype('int64')
merged_df.sort_values(by='Country',ascending=False,inplace=True)
merged_df.reset_index(drop=True,inplace=True)
```
